```
import pymongo

client = pymongo.MongoClient("mongodb+srv://testuser:beam4456@cluster0.ntixl.mongodb.net/Demo1?retryWrites=true&w=majority")
db = client.student_scores
```

```
file = open('score.csv')
student_scores=[]
for line in file:
  id,quiz1,quiz2,quiz3,quiz4,quiz5 = line.split(',')
  student_scores.append({'id':id,
               'quiz 1':int(quiz1),
               'quiz 2':int(quiz2),
               'quiz 3':int(quiz3),
               'quiz 4':int(quiz4),
               'quiz 5':int(quiz5),
               'sum':(int(quiz1)+int(quiz2)+int(quiz3)+int(quiz4)+int(quiz5))
               })
```

```
for student in student_scores:
    db.scores.insert_one(student)
```

```
db.scores_many.insert_many(student_scores,ordered=False)
```
