```
import random

data=[random.randint(0,10) for i in range(50)]
freq=[0 for j in range(11)]

for n in range(len(data)):
  for m in range(len(freq)):
    if data[n] == m:
      freq[m] += 1

# print(sorted(data))

for i in range(11):
  print(str(i)+':',freq[i])
```

```
import random
from collections import Counter 

# data=[random.randint(1,1000) for i in range(50000)]
# print(Counter(data).most_common())

file=open('CH3Thailand Posts.csv','r')
lines=file.readlines()
file.close()

hash=[]
tokens=[line.strip('\n').split(' ') for line in lines]
for i in tokens:
  for j in i:
    if '#' in j:
      hash.append(j)

print(Counter(hash).most_common())
```
