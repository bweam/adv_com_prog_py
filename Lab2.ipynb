```
import math

d = input('Enter degree: ')
ans1 = int(d)/180*math.pi
ans2 = math.radians(int(d))
print('From equation:', ans1)
print('From math.radians:', ans2)
```

```
sen = input('Input sentence: ')
ans = ''
for c in sen:
    if c.isalpha() or c.isdigit():
        ans += c
print('Remove punctuations:', ans)
print('Total characters:', len(ans))
```

```
file = open('score.csv')
sum = 0
temp = 0
for line in file:
  data = line.split(',')
  temp += 1
  if temp == 1: #negate the first row containing columns' headings
    continue
  tempsum = 0
  for i in range(1,6):
    tempsum += int(data[i])
  if tempsum > sum:
    sum = tempsum
print('Highest total score is',sum)
```

```
file = open('pm25.csv')
high = 0
date = ''
for line in file:
  dt,pm25 = line.strip().split(',')
  if pm25 != '-':
    if int(pm25) > high:
      high = int(pm25)
      date = dt
print('The highest pm2.5 is', high)
print('The highest date/time is', date)
```
