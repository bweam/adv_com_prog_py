###While

```
sum = 0
n = 1
data = int(input('Enter data ' + str(n) + ': '))
while data >= 0:
  sum += data
  n += 1
  data = int(input('Enter data ' + str(n) + ': '))
if n > 1:
  print(sum/(n-1))
else:
  print('No data')
```

###Selection

```
score = 167
if score > 100 or score < 10:
  grade = 'Error'
else:
  if score >= 80:
    grade = 'A'
  elif score >= 70:
    grade = 'B'
  elif score >= 60:
    grade = 'C'
  elif score >= 50:
    grade = 'D'
  else:
    grade = 'F'
print(grade)
```
