# Problem 2

```
def get_color(x):
  if x >= 50:
    return 'red'
  elif x >= 37:
    return 'yellow'
  return 'green'

pm_df=pd.read_csv('pm25.csv')
pm_df.columns=['dt','pm']
pm_df=pm_df[pm_df['pm'] != '-']
pm_df['pm']=pd.to_numeric(pm_df['pm'])
pm_df['dt']=pd.to_datetime(pm_df['dt'])
pm_df['hr']=pm_df['dt'].dt.hour

colors = [get_color(x) for x in list(pm_df.groupby('hr').mean()['pm'])]

sbs.barplot(data=pm_df,x='hr',y='pm',palette=colors)
```

# Problem 3

```
hp_df = pd.read_csv('happiness.csv') #world happiness
```

```
hp_df
```

```
hp_df.rename(columns={'Score':'Happiness'},inplace=True)
```

World Happiness ranking

```
import matplotlib.pyplot as plt

def get_color2(x):
  if x >= 7:
    return 'green'
  if x >= 6:
    return 'yellow'
  if x >= 5:
    return 'orange'
  if x >= 4:
    return 'red'
  return 'black'

colors2 = [get_color2(x) for x in hp_df['Happiness']]
fig,ax=plt.subplots(figsize=(24,8))
chart=sbs.barplot(x='Country or region', y='Happiness',data=hp_df,palette=colors2)
chart.set_xticklabels(chart.get_xticklabels(),rotation=90)
```

Does people's perceptions of corruption correlates to people's level of happiness

```
import plotly.express as px

fig=px.scatter(hp_df,x='Perceptions of corruption',y='Happiness',color='Country or region')
fig.show()
```
